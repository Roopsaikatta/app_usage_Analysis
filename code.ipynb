```python
df = pd.read_csv('app_usage.csv')
```

Convert 'Date' to datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
```

 Combine 'Date' and 'Time' into a single datetime column for easier analysis

```python
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'])
```

Convert 'Duration' to a timedelta object

```python
df['Duration'] = pd.to_timedelta(df['Duration'])
```

Create additional columns for analysis

```python
df['DayOfWeek'] = df['DateTime'].dt.day_name()
df['HourOfDay'] = df['DateTime'].dt.hour
df['TotalDurationSeconds'] = df['Duration'].dt.total_seconds()
```

```python
total_usage_per_app = df.groupby('App')['TotalDurationSeconds'].sum().sort_values(ascending=False)
```

```python
usage_by_day = df.groupby('DayOfWeek')['TotalDurationSeconds'].sum().sort_values(ascending=False)
```
